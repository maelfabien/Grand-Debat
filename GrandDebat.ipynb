# Average answer rate

```python
answer = np.array(pub.count())
no_answer = np.array(pub.isna().sum())
tot = answer + no_answer
avg_answer = answer/tot
i = np.arange(0, len(avg_answer))
plt.figure(figsize=(12,8))
plt.bar(i, avg_answer)
plt.title("Taux moyen de réponse par question")
plt.show()
```
